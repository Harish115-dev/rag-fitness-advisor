```python
import pypdf
def load_pdf_by_page(path):
    pages = []
    reader = pypdf.PdfReader(path)

    for page_num, page in enumerate(reader.pages, start=1):
        text = page.extract_text()
        if text:
            pages.append({
                "page": page_num,
                "text": text.strip()
            })

    return pages
```

```python
#chunker
```

```python
def chunker(
    pages,
    headings,
    source_name,
    max_words=160
):
    chunks = []
    chunk_id = 0

    for page in pages:
        current_section = "General"
        buffer = []
        word_count = 0

        for line in page["text"].split("\n"):
            line = line.strip()
            if not line:
                continue

            # ðŸ”¹ heading detection
            if line in headings:
                if buffer:
                    chunks.append({
                        "chunk_id": chunk_id,
                        "source": source_name,
                        "page": page["page"],
                        "section": current_section,
                        "text": " ".join(buffer)
                    })
                    chunk_id += 1

                current_section = line
                buffer = []
                word_count = 0
                continue

            buffer.append(line)
            word_count += len(line.split())

            if word_count >= max_words:
                chunks.append({
                    "chunk_id": chunk_id,
                    "source": source_name,
                    "page": page["page"],
                    "section": current_section,
                    "text": " ".join(buffer)
                })
                chunk_id += 1
                buffer = []
                word_count = 0

        
        if buffer:
            chunks.append({
                "chunk_id": chunk_id,
                "source": source_name,
                "page": page["page"],
                "section": current_section,
                "text": " ".join(buffer)
            })
            chunk_id += 1

    return chunks
```

```python
def merge_chunks(datasets):
    final_chunks = []
    chunk_id = 0

    for dataset in datasets:
        for chunk in dataset:
            final_chunks.append({
                "chunk_id": chunk_id,
                "source": chunk["source"],
                "page": chunk["page"],
                "section": chunk["section"],
                "text": chunk["text"]
            })
            chunk_id += 1

    return final_chunks
```
