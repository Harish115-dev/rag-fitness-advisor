```python
import pypdf
def load_pdf_by_page(path):
    pages = []
    reader = pypdf.PdfReader(path)

    for page_num, page in enumerate(reader.pages, start=1):
        text = page.extract_text()
        if text:
            pages.append({
                "page": page_num,
                "text": text.strip()
            })

    return pages
```

```python
def chunker(pages, headings, source_name, max_words=160):
    chunks = []
    chunk_id = 0
    current_section = "General"
    buffer = []
    word_count = 0
    
    for page in pages:
        for line in page["text"].split("\n"):
            line = line.strip()
            if not line:
                continue
            
            if any(line.startswith(h) for h in headings):
                if buffer:
                    chunks.append({
                        "chunk_id": chunk_id,
                        "source": source_name,
                        "page": page["page"],
                        "section": current_section,
                        "text": " ".join(buffer)
                    })
                    chunk_id += 1
                current_section = line
                buffer = []
                word_count = 0
                continue
            
            buffer.append(line)
            word_count += len(line.split())
            
            if word_count >= max_words:
                chunks.append({
                    "chunk_id": chunk_id,
                    "source": source_name,
                    "page": page["page"],
                    "section": current_section,
                    "text": " ".join(buffer)
                })
                chunk_id += 1
                buffer = []
                word_count = 0
    
    if buffer:
        chunks.append({
            "chunk_id": chunk_id,
            "source": source_name,
            "page": page["page"],
            "section": current_section,
            "text": " ".join(buffer)
        })
    
    return chunks
```

```python
import re
def activity_compendium_chunker(pages):
    chunks = []
    chunk_id = 0
    pattern = re.compile(
        r'^([A-Za-z &]+)\s+(\d{5})\s+([\d.]+)\s+(.*)$'
    )

    for page in pages:
        for line in page["text"].split("\n"):
            line = line.strip()

            if not line:
                continue

            match = pattern.match(line)
            if not match:
                continue  

            category, code, met, description = match.groups()

            chunks.append({
                "chunk_id": chunk_id,
                "doc_type": "activity_compendium",
                "category": category.strip(),
                "activity_code": code,
                "met": float(met),
                "activity": description.strip(),
                "page": page["page"],
                "source": "2024 Adult Compendium of Physical Activities"
            })

            chunk_id += 1

    return chunks
```

```python
compendium_chunks=activity_compendium_chunker(compendiam_pages)
```

```python
with open("data/compendiumchunks.json", "w", encoding="utf-8") as f:
        json.dump(compendium_chunks, f, indent=2, ensure_ascii=False)
```
