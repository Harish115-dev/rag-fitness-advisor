```python
import json
```

```python
def exercise_to_text(ex):
    return f"""
    Exercise Name: {ex.get("name")}
    Category: {ex.get("category")}
    Difficulty Level: {ex.get("level")}
    Force Type: {ex.get("force")}
    Mechanic: {ex.get("mechanic")}
    Equipment Required: {ex.get("equipment")}
    Primary Muscles: {", ".join(ex.get("primaryMuscles", []))}
    Secondary Muscles: {", ".join(ex.get("secondaryMuscles", []))}
    How to Perform:
    {" ".join(ex.get("instructions", []))}
    """.strip()
```

```python
def chunk_exercises():
    with open("data/json/exercises.json","r",encoding="utf-8") as data:
        exercises=json.load(data)
    
    chunks=[]
    
    for ex in exercises:
         chunk = {
            "content": exercise_to_text(ex),
            "metadata": {
                "source": "exercises.json",
                "exercise_id": ex.get("id"),
                "name": ex.get("name"),
                "level": ex.get("level"),
                "equipment": ex.get("equipment"),
                "category": ex.get("category"),
                "primary_muscles": ex.get("primaryMuscles", [])
            }
        }
         chunks.append(chunk)
    with open("data/exercise_chunks.json","w", encoding="utf-8") as f:
         json.dump(chunks, f, indent=2)
        
        
```

```python
chunk_exercises()
```
